### 4)The LookupError class in Python is a base class for exceptions that occur when a key or index used to access a container is invalid or not found.

```python
my_dict = {'a': 1, 'b': 2, 'c': 3}
try:
    value = my_dict['d']  
except KeyError as e:
    print("Error:", e)
```

```python
my_list = [1, 2, 3, 4, 5]
try:
    value = my_list[10]
except IndexError as e:
    print("Error:", e)
```
